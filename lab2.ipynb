```python
import math

def bezwzgl_aproksym_a_przy_pomocy_b(a, b):
    print("Błąd bezwzględny aproksymacji "+str(a)+" przy pomocy "+str(b)+" jest równy "+ str(abs(a-b)))
    return abs(a-b)

bezwzgl_aproksym_a_przy_pomocy_b(math.pi, 22/7)
bezwzgl_aproksym_a_przy_pomocy_b(math.exp(1), 2.718)
bezwzgl_aproksym_a_przy_pomocy_b(pow(10, math.pi), 1397)
bezwzgl_aproksym_a_przy_pomocy_b(math.factorial(9), math.sqrt(18*math.pi)*(pow(9/math.exp(1), 9)))

def wzgl_aproksym_a_przy_pomocy_b(a, b):
    print("Błąd względny aproksymacji "+str(a)+" przy pomocy "+str(b)+" jest równy "+ str(abs(a-b)/a))
    return abs(a-b)/a

wzgl_aproksym_a_przy_pomocy_b(math.pi, 22/7)
wzgl_aproksym_a_przy_pomocy_b(math.exp(1), 2.718)
wzgl_aproksym_a_przy_pomocy_b(pow(10, math.pi), 1397)
wzgl_aproksym_a_przy_pomocy_b(math.factorial(9), math.sqrt(18*math.pi)*(pow(9/math.exp(1), 9)))
    
```

***Zadanie 2.***  

```python
import numpy as np
def ciag(n):
    rozwiazania=[]
    for p in range(1, n+1):
        rozwiazania.append((pow(10, p)*(1+p*np.round(math.pi, 15)*pow(10, -p))-pow(10, p))/p)
    
    return rozwiazania

print(ciag(20))
```

Ad. 1
Wartość G(p) w sposób analityczny nie zależy od p. Powinno zwracać, za każdym razem np.round(math.pi, 15). 

```python
#Ad. 2
wektor = ciag(20)
bezwzgl = []
wzgl=[]
for x in range (1, 21):
    bezwzgl.append(bezwzgl_aproksym_a_przy_pomocy_b(np.round(math.pi, 15),wektor[x-1]))
    wzgl.append(wzgl_aproksym_a_przy_pomocy_b(np.round(math.pi, 15), wektor[x-1]))
    
print(bezwzgl)
print(wzgl)
```

```python
#Ad. 2 cd.
import matplotlib.pyplot as plt
x=[]
for i in range(1, 21):
    x.append(i)

plt.semilogy(x,wzgl,label="Błędy względne")
plt.semilogy(x,bezwzgl,label="Błędy bezwzględne")
plt.title('Błędy względne i bezwzględne')
plt.ylabel('Wartość błędu')
plt.xlabel('Krok')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

plt.plot(x,wzgl,label="Błędy względne")
plt.plot(x,bezwzgl,label="Błędy bezwzględne")
plt.title('Błędy względne i bezwzględne')
plt.ylabel('Wartość błędu')
plt.xlabel('Krok')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
